```python
second_to_seq = np.zeros(3600*24)
second_to_n = np.zeros(3600*24)

trace = '/nfs/memc_200m_shift.tr.ant'
with open(trace, 'r') as fin:
    lines = fin.readlines()
    i = 0
    cnt_n = 0
    start_t = 0
    for line in lines:
        t = int(line.split()[1]) // 1000000
        if start_t == 0:
            start_t = t
            cur_t = t
        if cur_t == t:
            cnt_n += 1
        else:
            second_to_seq[t - start_t] = i
            second_to_n[t - start_t] = cnt_n
            cnt_n = 0
            cur_t = t
        i += 1
    second_to_seq = second_to_seq[:t]
    second_to_n = second_to_n[:t]

print(second_to_seq[-10:])
print(second_to_n[-10:])
```

```python
import math
def rolling_sum(n) :
    ret = np.cumsum(second_to_n, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret

def remove_repeat(a, n):
    index = []
    for i in range(len(a)):
        insert = 1
        for k in index:
            if abs(k-a[len(a)-1-i]) < n:
                insert = 0
                break
        if insert:
            index.append(a[len(a)-1-i])
    return np.array(index)

win = 10
rolling_n = rolling_sum(win)
indices = np.argsort(rolling_n)[-100:].astype(int)
print(len(rolling_n))
indices = remove_repeat(indices, win)

print(indices)
print(second_to_seq[indices-win])
print(second_to_seq[indices])
#print(np.sum(n_per_seconds[indices[0]-9: indices[0]+1]))
print(rolling_n[indices])

e2e_latencys = [t_ine, t_mat, t_lru, t_1mat]
misses = [m_ine, m_mat, m_lru, m_1mat]
e2e_peak = [[], [], [], []]
miss_peak = [[], [], [], []]
latency_peak_index = []
for i in range(4):
    peak_index = []
    for j in range(len(indices)):
        #print(int(second_to_seq[indices[j]-1]), int(second_to_seq[indices[j]]))
        #peak_index.extend(range(int(second_to_seq[indices[j]-1]), int(second_to_seq[indices[j]])))
        #    print(indices[j] * 100000 - 300000, indices[j] * 100000 + 200000)
        e2e_peak[i].extend(e2e_latencys[i][int(second_to_seq[indices[j]-win]): int(second_to_seq[indices[j]])])
        miss_peak[i].append(np.average(misses[i][int(second_to_seq[indices[j]-win*2]): int(second_to_seq[indices[j]])]))
        for k in range(int(second_to_seq[indices[j]-win]) // 100000 + 1, math.ceil(second_to_seq[indices[j]] / 100000) + 1):
            if i == 0:
                latency_peak_index.append(k)
        #print(peak_index)
    #e2e_peak[i] = e2e_latencys[i][peak_index]
    e2e_peak[i] = np.array(e2e_peak[i])
    e2e_peak[i] = np.sort(e2e_peak[i])
print('done')
latency_peak_index = [*set(latency_peak_index)]
print(latency_peak_index)
print(len(e2e_peak[0]))
```
